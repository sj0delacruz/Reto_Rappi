```python
from glob import glob
import numpy as np

labels_names=np.loadtxt('/content/data/dataset0/labels/labels.txt',str)
root_labels_path = '/content/yolov5/runs/detect/exp2/labels/' #roca validar
labels_paths = glob(root_labels_path+'*.txt')
for labels_path in labels_paths:
  targets = np.loadtxt(labels_path, ndmin=2)
  img_path=labels_path.replace('yolov5/runs/detect/exp2/labels', 'data/dataset0/images/valid').replace('txt', 'png')
  img = Image.open(img_path)
  for n, i in enumerate(targets):
    img_cropped=img.crop( ((i[1]-i[3]/2)*img.width, (i[2]-i[4]/2)*img.height, (i[1]+i[3]/2)*img.width, (i[2]+i[4]/2)*img.height) )
    img_cropped.save(img_path.replace('dataset0/images/valid', 'dataset1/images/train_augmented').replace('.png', '_'+labels_names[n]+'.png'))
```
